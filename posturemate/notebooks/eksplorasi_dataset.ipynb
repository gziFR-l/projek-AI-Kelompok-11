# Eksplorasi Dataset PostureMate

Notebook ini digunakan untuk memeriksa dataset postur tubuh, mengecek kualitas data, dan menyiapkan langkah analisis lebih lanjut.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

print('Libraries imported successfully.')
```

```python
# Load Dataset Files
base_dir = Path('..')
main_csv = base_dir / 'data' / 'data.csv'
original_csv = base_dir / 'data' / 'dataset_original.csv'

print('Main dataset path exists:', main_csv.exists())
print('Original dataset path exists:', original_csv.exists())

main_df = pd.read_csv(main_csv) if main_csv.exists() else None
original_df = pd.read_csv(original_csv) if original_csv.exists() else None
```

```python
# Inspect Dataset Shape and Columns
if main_df is not None:
    print('Main dataset shape:', main_df.shape)
    print('Main columns:', list(main_df.columns))

if original_df is not None:
    print('Original dataset shape:', original_df.shape)
    print('Original columns:', list(original_df.columns))
```

```python
# Check Missing Values and Data Types
if main_df is not None:
    print('Missing values in main dataset:')
    print(main_df.isnull().sum())
    print('\nData types:')
    print(main_df.dtypes)
```

```python
# Explore Posture Class Distribution
if main_df is not None and 'label' in main_df.columns:
    counts = main_df['label'].value_counts()
    print(counts)
    plt.figure(figsize=(8, 4))
    sns.countplot(data=main_df, x='label')
    plt.title('Distribusi Kelas Postur')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Visualize Feature Relationships
if main_df is not None:
    numeric_cols = main_df.select_dtypes(include=[np.number]).columns.tolist()
    print('Numeric columns:', numeric_cols)
    if len(numeric_cols) >= 2:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=main_df, x=numeric_cols[0], y=numeric_cols[1], hue='label', alpha=0.8)
        plt.title('Hubungan Fitur Antar Kelas')
        plt.show()
```

```python
# Summarize Statistical Insights
if main_df is not None:
    print(main_df.describe())
```

```python
# Prepare Dataset for Future Modeling
if main_df is not None:
    cleaned_df = main_df.copy()
    cleaned_df = cleaned_df.dropna()
    print('Cleaned dataset shape:', cleaned_df.shape)
    print('Next steps: feature engineering, model training, and evaluation.')
```
